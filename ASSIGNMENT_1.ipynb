```python
  # Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/retail_sales_dataset.csv")
```

```python
# Univariate Analysis

# 1. Transaction ID: Count of unique transaction IDs
print("\nUnique Transaction IDs:", df['Transaction ID'].nunique())
```

```python
# 2. Date: Frequency of transactions over time
df['Date'] = pd.to_datetime(df['Date'])  # Convert to datetime format
plt.figure(figsize=(12, 6))
df['Date'].value_counts().sort_index().plot(kind='line', title='Transaction Frequency Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Transactions')
plt.show()
```

```python
# 3. Customer ID: Count of unique customers
print("\nUnique Customer IDs:", df['Customer ID'].nunique())
```

```python
# 4. Gender: Distribution of male vs. female customers
plt.figure(figsize=(6, 4))
df['Gender'].value_counts().plot(kind='bar', title='Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
#You can see no.s of males and females is almost similar
```

```python
# 5. Age: Distribution of customers ages
plt.figure(figsize=(8, 4))
df['Age'].plot(kind='hist', bins=10, title='Age Distribution', color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
#as you can see that age distribution is quite balanced and mostly age is concerntarted around 40 to 50
```

```python
# 6. Product Category: Distribution of product categories
plt.figure(figsize=(10, 5))
df['Product Category'].value_counts().plot(kind='bar', title='Product Category Distribution', color='lightcoral')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
#count is highest of clothing followed by electronics than least beauty
```

```python
# 7. Quantity: Distribution of quantities purchased
plt.figure(figsize=(8, 4))
sns.histplot(df['Quantity'], bins=10, kde=True, color='green')
plt.title('Distribution of Quantity Purchased')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
#quantity purchased is quite balanced
```

```python
# 8. Price per Unit: Distribution of price per unit
plt.figure(figsize=(8, 4))
sns.histplot(df['Price per Unit'], bins=20, kde=True, color='orange')
plt.title('Distribution of Price per Unit')
plt.xlabel('Price per Unit')
plt.ylabel('Frequency')
plt.show()
#as you can see distribution is  not that balanced
```

```python
# 9. Total Amount: Distribution of the total amount of transactions
plt.figure(figsize=(8, 4))
sns.histplot(df['Total Amount'], bins=20, kde=True, color='purple')
plt.title('Distribution of Total Amount of Transactions')
plt.xlabel('Total Amount')
plt.ylabel('Frequency')
plt.show()
#the distribution is quite right skewed
#total amount of transaction is concerntrated around less amount
```

```python
# Bivariate Analysis

# 1. Gender vs. Total Amount
plt.figure(figsize=(6, 4))
sns.boxplot(x='Gender', y='Total Amount', data=df)
plt.title('Total Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Amount')
plt.show()
#there are no significant outliers for 'gender'
```

```python
# 2. Age vs. Total Amount
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Total Amount', data=df, hue='Gender')
plt.title('Age vs Total Amount')
plt.xlabel('Age')
plt.ylabel('Total Amount')
plt.show()
#again you can see it is more around less amount
```

```python
# 3. Product Category vs. Total Amount
plt.figure(figsize=(12, 6))
sns.boxplot(x='Product Category', y='Total Amount', data=df, palette='Set2')
plt.title('Total Amount by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.show()
#you can see amount by category is quite less for clothing even though its no. is quite high
#moreover, 2 outliers can be seen for the clothing part
```

```python
# 4. Quantity vs. Total Amount
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Quantity', y='Total Amount', data=df, hue='Product Category')
plt.title('Quantity vs Total Amount')
plt.xlabel('Quantity')
plt.ylabel('Total Amount')
plt.show()
```

```python
# 5. Price per Unit vs. Total Amount
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Price per Unit', y='Total Amount', data=df, hue='Product Category')
plt.title('Price per Unit vs Total Amount')
plt.xlabel('Price per Unit')
plt.ylabel('Total Amount')
plt.show()
```

```python

# Filter the dataframe to include only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Correlation Heatmap using only numeric columns
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
#more of negative correlations can be seen between attributes which means if one increases, the other decreases
#however, some +ve correlations can also be seen between some attributes
```
